```python
from matplotlib.patches import Patch
def plot_nice(new_transform, y, title, max):
  ##this ensures consistent plots for visualization
  legend_elements = [Patch(fc = 'xkcd:dark lavender', label=f'Less than {0.2*max:.0f} Points'), Patch(fc = 'xkcd:sky blue', label=f'{0.2*max:.0f}-{0.4*max-1:.0f} Points'),
                     Patch(fc = 'xkcd:sage', label=f'{0.4*max:.0f}-{0.6*max-1:.0f} Points'), Patch(fc = 'xkcd:lemon', label=f'{0.6*max:.0f}-{0.8*max-1:.0f} Points'),
                     Patch(fc = 'xkcd:coral', label=f'{0.8*max:.0f}-{max-1} Points'), Patch(fc = 'xkcd:pink', label=f'{max} or more Points')]

  plt.scatter(new_transform[(y<(0.2*max)).nonzero()[0], 0], new_transform[(y<(0.2*max)).nonzero()[0], 1], c = 'xkcd:dark lavender')
  plt.scatter(new_transform[(((0.2 *max)<=y) & (y<(0.4 *max))).nonzero()[0], 0], new_transform[(((0.2 *max)<=y) & (y<(0.4 *max))).nonzero()[0], 1], c = 'xkcd:sky blue')
  plt.scatter(new_transform[(((0.4 *max)<=y) & (y<(0.6 *max))).nonzero()[0], 0], new_transform[(((0.4 *max)<=y) & (y<(0.6 *max))).nonzero()[0], 1], c = 'xkcd:sage')
  plt.scatter(new_transform[(((0.6 *max)<=y) & (y<(0.8 *max))).nonzero()[0], 0], new_transform[(((0.6 *max)<=y) & (y<(0.8 *max))).nonzero()[0], 1], c = 'xkcd:lemon')
  plt.scatter(new_transform[(((0.8 *max)<=y) & (y<max)).nonzero()[0], 0], new_transform[(((0.8 *max)<=y) & (y<max)).nonzero()[0], 1], c = 'xkcd:coral')
  plt.scatter(new_transform[(y>=max).nonzero()[0], 0], new_transform[(y>=max).nonzero()[0], 1], c = 'xkcd:pink')
  plt.legend(handles = legend_elements)
  plt.title(title)

  plt.show()
```

```python
from sklearn.model_selection import train_test_split

#split into test and train
X_full_train, X_full_val, y_full_train, y_full_val = train_test_split(X_full_relevant, y_full, test_size=0.15, random_state=42)

X_pp_train, X_pp_val, y_pp_train, y_pp_val = train_test_split(X_pp_relevant, y_pp, test_size=0.15, random_state=42)

X_s_train, X_s_val, y_s_train, y_s_val = train_test_split(X_s_relevant, y_s, test_size=0.15, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

#full strength model
min = 100
scores = []
depths = np.arange(2, 20)
estimators = np.arange(20, 70)

for d in depths:
  for e in estimators:
    fullS_rf_reg = RandomForestRegressor(n_estimators=e, max_depth=d, random_state=42).fit(X_full_train, y_full_train[:,0])

    y_val_pred = fullS_rf_reg.predict(X_full_val)
    this_loss = mean_squared_error(y_full_val, y_val_pred)
    scores.append(this_loss)

    if this_loss < min:
      min = this_loss
      params = [d, e]

print(f"Min: {min}")
print(f"Max Depth: {params[0]}\nn_est: {params[1]}")
```

```python
#powerplay model

min = 100
depths = np.arange(2, 20)
estimators = np.arange(30, 100)

for d in depths:
  for e in estimators:
    PP_rf_reg = RandomForestRegressor(n_estimators=e, max_depth=d, random_state=42).fit(X_pp_train, y_pp_train[:,0])

    y_val_pred = PP_rf_reg.predict(X_pp_val)
    this_loss = mean_squared_error(y_pp_val, y_val_pred)

    if this_loss < min:
      min = this_loss
      params = [d, e]

print(f"Min: {min}")
print(f"Max Depth: {params[0]}\nn_est: {params[1]}")
```

```python
#shorthanded model
min = 100
depths = np.arange(2, 20)
estimators = np.arange(30, 100)

for d in depths:
  for e in estimators:
    s_rf_reg = RandomForestRegressor(n_estimators=e, max_depth=d, random_state=42).fit(X_s_train, y_s_train[:,0])

    y_val_pred = s_rf_reg.predict(X_s_val)
    this_loss = mean_squared_error(y_s_val, y_val_pred)

    if this_loss < min:
      min = this_loss
      params = [d, e]

print(f"Min: {min}")
print(f"Max Depth: {params[0]}\nn_est: {params[1]}")
```

```python
class HockeyDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X.copy()).float()
        self.y = torch.from_numpy(y.copy()).float()
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
# # Note: No metric needed and remember to change the loss function

def train_and_validate(train_loader, val_loader, model, optimizer,
                      criterion, num_epochs, metric=None):
  history = {
        'epoch': [],
        'train_loss': [],
        'train_metric': [],
        'val_loss': [],
        'val_metric': []
    }  # Initialize a dictionary to store results from each epoch

  for epoch in range(num_epochs):
        model.train()  # Set the model to training mode
        epoch_loss = 0.0  # Initialize the epoch loss and metric values
        epoch_metric = 0.0

        # Training loop
        for X, y in train_loader:
            optimizer.zero_grad()  # Clear existing gradients
            outputs = model(X)  # Make predictions
            outputs = outputs[:,0]
            loss = criterion(outputs, y)  # Compute the loss
            loss.backward()  # Compute gradients
            optimizer.step()  # Update model parameters

            epoch_loss += loss.item()
            if metric is not None:
                epoch_metric += metric(outputs, y)
            else:
                epoch_metric += 0

        # Average training loss and metric
        epoch_loss /= len(train_loader)
        epoch_metric /= len(train_loader)

        # Validation loop
        model.eval()  # Set the model to evaluation mode
        with torch.no_grad():  # Disable gradient calculation
            val_loss = 0.0
            val_metric = 0.0
            for X_val, y_val in val_loader:
                outputs_val = model(X_val)  # Make predictions
                outputs_val = outputs_val[:,0]
                val_loss += criterion(outputs_val, y_val).item()  # Compute loss
                if metric is not None:
                    val_metric += metric(outputs_val, y_val)
                else:
                    val_metric += 0

            val_loss /= len(val_loader)
            val_metric /= len(val_loader)

        # Append epoch results to history
        history['epoch'].append(epoch)
        history['train_loss'].append(epoch_loss)
        history['train_metric'].append(epoch_metric)
        history['val_loss'].append(val_loss)
        history['val_metric'].append(val_metric)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {epoch_loss:.4f}, '
              f'Train Metric: {epoch_metric:.4f}, Val Loss: {val_loss:.4f}, '
              f'Val Metric: {val_metric:.4f}')
  return history, model

```

```python
def test_model(model, data_loader, criterion):
  model.eval()  # Set the model to evaluation mode

  total_loss = 0.0  # Initialize the total loss

  with torch.no_grad():  # Disable gradient tracking
      for batch in data_loader:
          X, y = batch

          # Pass the data to the model and make predictions
          outputs = model(X)
          outputs = outputs[:,0]

          # Compute the loss
          loss = criterion(outputs, y)

          # Add the loss and metric for the batch to the total values
          total_loss += loss.item()

  # Average loss and metric for the entire dataset
  avg_loss = total_loss / len(data_loader)


  print(f'Test Loss: {avg_loss:.4f}')

  return avg_loss
```
